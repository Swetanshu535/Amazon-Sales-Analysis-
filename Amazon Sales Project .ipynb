```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
df=pd.read_csv("C:\\Users\\Swetanshu\\Desktop\\Amazon_sales 2025.csv")
print(df)
```

What are the monthly sales trends throughout the year?

Graph type: Line chart or bar chart

Purpose: To visualize seasonal or monthly sales fluctuations.

```python
df['Date'] = pd.to_datetime(df['Date'])


df['Month'] = df['Date'].dt.month_name()


month_order = ['January', 'February', 'March', 'April', 'May', 'June', 
               'July', 'August', 'September', 'October', 'November', 'December']


monthly_sales = df.groupby('Month')['Total Sales'].sum().reindex(month_order)


plt.figure(figsize=(12,6))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values, marker='o')
plt.title('Monthly Sales Trend - 2025')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

#Which product categories generate the highest total sales?
#•	Graph type: Horizontal bar chart or pie chart
#•	Purpose: Identify top-selling categories.

category_sales = df.groupby('Category')['Total Sales'].sum().sort_values(ascending=False)


plt.figure(figsize=(10,6))
sns.barplot(x=category_sales.values, y=category_sales.index, palette='viridis')
plt.title('Total Sales by Product Category')
plt.xlabel('Total Sales')
plt.ylabel('Product Category')
plt.tight_layout()
plt.show()


```

```python
#How does the number of reviews relate to sales volume?

#Graph type: Scatter plot

#Purpose: To check if more reviews lead to higher sales.

print(df.columns)


plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Status', y='Total Sales', hue='Category', palette='tab10', alpha=0.7)
plt.title('Relationship Between Number of Reviews and Sales Volume')
plt.xlabel('Number of Reviews')
plt.ylabel('Sales Volume')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#What is the distribution of product ratings across all products?
#•	Graph type: Histogram or box plot
#•	Purpose: Understand the rating quality of sold products.



plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 2)
sns.boxplot(x=df['Payment Method'], color='lightgreen')
plt.title('Box Plot of Product Ratings')
plt.xlabel('Payment Method')

plt.tight_layout()
plt.show()
```

```python
#Which regions/states/cities contribute the most to total sales?
#•	Graph type: Geographical heat map or bar chart
#•	Purpose: Pinpoint top-performing locations.

sales_by_city = df.groupby('Product')['Price'].sum().sort_values(ascending=False).head(10)


plt.figure(figsize=(12, 6))
sns.barplot(x=sales_by_city.values, y=sales_by_city.index, palette='viridis')
plt.title('Top 10 Cities by Total Sales')
plt.xlabel('Product')
plt.ylabel('Price')
plt.tight_layout()
plt.show()
```
